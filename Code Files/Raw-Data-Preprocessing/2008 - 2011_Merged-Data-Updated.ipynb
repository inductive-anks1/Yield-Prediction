### 7. Himachal Pradesh

```python
data_HP.head()
```

```python
import numpy as np

data_HP = data_HP.rename(columns={np.nan: 'Village'})
```

```python
data_HP.shape
```

```python
def make_unique(cols):
    seen = {}
    for i, col in enumerate(cols):
        if col not in seen:
            seen[col] = 1
        else:
            seen[col] += 1
            cols[i] = f"{col}_{seen[col]}"
    return cols

# Apply the function to the columns
data_HP.columns = make_unique(list(data_HP.columns))
```

```python
data_HP.head()
```

```python
data_HP.shape
```

```python
village_rows = data_HP.copy()
village_2_rows = data_HP.copy()

village_rows['CLUSTER VILLAGES'] = village_rows['Village']
village_2_rows['CLUSTER VILLAGES'] = village_2_rows['Village_2']

new_df = pd.concat([data_HP, village_rows, village_2_rows], ignore_index=True)

new_df = new_df.drop(columns=['Village', 'Village_2'])

new_df = new_df.sort_values(by=['ZONE', 'TEHSIL  CODE', 'CLUSTER VILLAGES']).reset_index(drop=True)
```
