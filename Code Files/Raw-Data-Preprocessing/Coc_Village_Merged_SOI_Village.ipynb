```python
df_village_SOI['Village Name - COC'] = df_village_SOI['Village Name - COC'].str.lower()
df_village_localisation['Village Name'] = df_village_localisation['Village Name'].str.lower()
```

```python
def remove_leading_numbers(village_name):
    parts = village_name.split('. ', 1)  # Split into two parts: before and after the first '. '
    if len(parts) == 2 and parts[0].isdigit():  # Check if the first part is a digit
        return parts[1]
    return village_name


df_village_localisation['Village Name'] = df_village_localisation['Village Name'].apply(remove_leading_numbers)
```

```python
def remove_text_before_dot(village_name):
    return village_name.split('.', 1)[-1]  # Split on the first dot and take the part after it

df_village_localisation['Village Name'] = df_village_localisation['Village Name'].apply(remove_text_before_dot)
```

```python
df_village_localisation['Village Name'] = df_village_localisation['Village Name'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
```

```python
merged_df = pd.merge(df_village_localisation, df_village_SOI, how='inner', left_on='Village Name', right_on='Village Name - COC')
```

```python
import pandas as pd
from nltk.metrics.distance import jaro_similarity
```

```python
results_df = pd.DataFrame(columns=["Village Name", "score", "Matched Village"])

for village_1 in non_matching_village_localisation['Village Name'].unique():
    max_score = 0
    best_match = ''
    for village_2 in non_matching_village_SOI['Village Name - COC'].unique():
        score = jaro_similarity(village_1, village_2)
        if score > max_score:
            max_score = score
            best_match = village_2
    
    temp_df = pd.DataFrame([{"Village Name": village_1, "score": max_score, "Matched Village": best_match}])
    results_df = pd.concat([results_df, temp_df], ignore_index=True)
```

```python
threshold = 0.90
results_df = results_df[results_df['score'] >= threshold]
```

```python
results_df = pd.DataFrame(columns=["State Name", "District Name", "Village Name (Village List)", "Score", "Village Name (Centroid List)"])

# Iterate over unique state and district combinations in the non-matching village list
for (state, district), group in non_matching_village_list.groupby(['State Name', 'District Name']):
    # Filter the centroid DataFrame for the same state and district
    centroid_group = non_matching_village_centroid[
        (non_matching_village_centroid['State Name Centroid'] == state) & 
        (non_matching_village_centroid['District Name Centroid'] == district)
    ]
    
    for village_1 in group['Village Name'].unique():
        max_score = 0
        best_match = ''
        
        for village_2 in centroid_group['vilnmcoc'].unique():
            score = jaro_similarity(village_1, village_2)
            
            if score > max_score:
                max_score = score
                best_match = village_2
        
        # Append the result to the results DataFrame
        temp_df = pd.DataFrame([{
            "State Name": state,
            "District Name": district,
            "Village Name (Village List)": village_1,
            "Score": max_score,
            "Village Name (Centroid List)": best_match
        }])
        
        results_df = pd.concat([results_df, temp_df], ignore_index=True)
```
