```python
import pandas as pd
import pathlib
```

```python
curr_dir = pathlib.Path().resolve()
print(curr_dir)
```

```python
home_dir = curr_dir.parent.parent
```

```python
df = pd.read_csv(home_dir.as_posix() + '/data/Final Data/only-check-similarity.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0', 'Similarity'],inplace=True)
```

```python
df.head()
```

```python
filtered_df = df.iloc[999:1500, :]
```

```python
filtered_df.head()
```

```python
filtered_df['NAME'] = filtered_df['NAME'].str.lower()
```

```python
filtered_df['vilnmcoc'] = filtered_df['vilnmcoc'].str.lower()
```

```python
import pandas as pd
from nltk.metrics.distance import jaro_similarity
```

```python
results_df = pd.DataFrame(columns=["Village Name 1", "score", "Village Name 2"])

# Iterate through each row in the DataFrame
for index, row in filtered_df.iterrows():
    village_1 = row['NAME']
    village_2 = row['vilnmcoc']
    score = jaro_similarity(village_1, village_2)
    
    # Store the results in the DataFrame
    temp_df = pd.DataFrame([{"Village Name 1": village_1, "score": score, "Village Name 2": village_2}])
    results_df = pd.concat([results_df, temp_df], ignore_index=True)
```

```python
threshold = 0.80
results_df = results_df[results_df['score'] >= threshold]
```
