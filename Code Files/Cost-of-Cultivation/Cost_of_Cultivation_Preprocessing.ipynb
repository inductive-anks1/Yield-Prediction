### 1. Finding a Unique Id column for the Dataset

```python
df.head(10)
```

```python
df['year'].unique()
```

```python
df = df[~df['year'].isin([2005, 2006, 2007])]
```

```python
df['year'].unique()
```

```python
df['UniqueID'] = df['cluster_weight'].astype(str) + df['sizegroup'].astype(str) + df['pps'].astype(str) + df['state_code'].astype(str) + df['dist_code'].astype(str) + df['sub_id'].astype(str) + df['crop_code'].astype(str) + df['year'].astype(str) + df['tehsilcultivator'].astype(str) + df['season'].astype(str)
```

```python
df['Extra Number'] = df.groupby(['tehsilcultivator', 'pps']).cumcount() + 1
```

```python
df['UniqueID'] = df['UniqueID'].astype(str) + df['Extra Number'].astype(str)
```

```python
df['UniqueID'] = df['UniqueID'].str.replace('.', '', regex=False)
```

```python
df['UniqueID'] = df['UniqueID'].str.replace('nan', '', regex=False)
```

```python
len(df['UniqueID'].unique())
```

```python
duplicate_ids = df['UniqueID'][df['UniqueID'].duplicated(keep=False)]

df_duplicates = df[df['UniqueID'].isin(duplicate_ids)]
```

### Finding Number fo Unique Tehsils (State Wise)

### Checking total number of Tehsils Year wise

```python
grouped = df.groupby(['State Name', 'year'])['Tehsil Code'].nunique().reset_index()
```

```python
total_unique_tehsils_per_year = grouped.groupby('year')['Tehsil Code'].sum().reset_index()
total_unique_tehsils_per_year.columns = ['year', 'Total Number of unique Tehsils']
```
