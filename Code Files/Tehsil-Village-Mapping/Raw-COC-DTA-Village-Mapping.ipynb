### Expanding the Year

```python
def expand_year_range(df):
    new_rows = []
    for index, row in df.iterrows():
        start_year, end_year = map(int, row['Year'].split('-'))
        for year in range(start_year, end_year + 1):
            new_row = row.copy()
            new_row['Year'] = str(year)
            new_rows.append(new_row)
    return pd.DataFrame(new_rows)


expanded_df = expand_year_range(df_tehsil_single_village)
```

```python
expanded_df.head()
```

```python
merged_df = pd.merge(df, expanded_df, on=['Tehsil Code', 'State Name', 'Year', 'District Name'], how='inner')
```

### All India Village Shapes from SOI data merged with Centroid Data using Within Function

```python
import pandas as pd
import geopandas as gpd
```

```python
df_centroid_file = gpd.read_file(home_dir.as_posix() + '/data/Centroid_File/coc_2008_2020_v1.shp')
df_centroid_file_same_centroid = pd.read_csv(home_dir.as_posix() + '/data/Final Data/different_vilage_same_centroid.csv')
```

```python
df_centroid_file.head()
```

```python
df_centroid_file_same_centroid.head(3)
```

```python
vilnmococ_values = df_centroid_file_same_centroid['vilnmcoc'].unique()

df_centroid_file_unique_centroid = df_centroid_file[~df_centroid_file['vilnmcoc'].isin(vilnmococ_values)]
```

```python
df_centroid_file_unique_centroid.shape
```

```python
df_centroid_file.shape
```

```python
df_centroid_file_same_centroid.shape
```

```python
def remove_initials(text):
    text = pd.Series(text).str.replace(r'\b\w\.\s*', '', regex=True)
    return text.str.strip()
```

```python
df_centroid_file_unique_centroid['vilnmcoc'] = df_centroid_file_unique_centroid['vilnmcoc'].apply(remove_initials)
```

```python
def remove_brackets(text):
    text = pd.Series(text).str.replace(r'\s*\(.*?\)\s*', '', regex=True)
    return text.str.strip()
df_centroid_file_unique_centroid['vilnmcoc'] = df_centroid_file_unique_centroid['vilnmcoc'].apply(remove_brackets)
```

```python
df_centroid_file_unique_centroid.to_csv(home_dir.as_posix() + '/data/Final Data/Centroid_file_Unique_Centroid.csv', index=False)
```

### Merging

```python
df_centroid_file_unique_centroid.head()
```

```python
df_soi_shapes = gpd.read_file(home_dir.as_posix() + '/data/all-villages-shapes/all-villages-shapes.shp')
```

```python
df_soi_shapes.shape
```

```python
df_soi_shapes.head()
```

```python
df_soi_shapes.head()
```

```python
df_soi_shapes['COC_Village'] = None

for i, centroid_row in df_centroid_file_unique_centroid.iterrows():
    point = centroid_row['geometry']
    COC_Village = centroid_row['vilnmcoc']
    
    if point is not None:
        for j, polygon_row in df_soi_shapes.iterrows():
            polygon = polygon_row['geometry']
            
            if polygon is not None and point.within(polygon):
                df_soi_shapes.at[j, 'COC_Village'] = COC_Village
```
