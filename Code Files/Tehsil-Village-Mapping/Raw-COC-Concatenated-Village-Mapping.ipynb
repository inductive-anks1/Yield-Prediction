```python
dataframe_list = [
    df_concat_1, df_concat_2, df_concat_3, df_concat_4, df_concat_5, 
    df_concat_6, df_concat_7, df_concat_8, df_concat_9, df_concat_10, 
    df_concat_11, df_concat_12
]

df_concat = pd.concat(dataframe_list, ignore_index=True)
```

```python
df_concat['state_name'] = df_concat['state_name'].replace('andhra pradesh ', 'andhra pradesh')
```

```python
df_concat['state_name'] = df_concat['state_name'].replace('himachal pradesh ', 'himachal pradesh')
```

```python
df_concat['state_name'] = df_concat['state_name'].replace('odisha', 'orissa')
```

```python
def expand_year_range(df):
    new_rows = []
    for index, row in df.iterrows():
        start_year, end_year = map(int, row['Year'].split('-'))
        for year in range(start_year, end_year):
            new_row = row.copy()
            new_row['Year'] = f"{year}-{str(year + 1)[-2:]}"
            new_rows.append(new_row)
    return pd.DataFrame(new_rows)

expanded_df = expand_year_range(df_tehsil_single_village)
```

```python
expanded_df.head()
```

```python
df_concat = df_concat.rename(columns={'state_name': 'State Name', 'tehsil_code': 'Tehsil Code', 'year':'Year'})
```

```python
df_concat = df_concat.astype(str)
```

```python
COC_village_merged_df = pd.merge(df_concat, expanded_df, on=['Tehsil Code', 'State Name', 'Year'], how='inner')
```
